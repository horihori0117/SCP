```python
# ✅ Google Colab 用：個体値の SCP ランキングツール（HTMLカード形式 + タイ順位対応）

import csv
import math
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import functools

# -----------------------------
# 📂 CSV 読み込み関数
# -----------------------------
def load_stats_from_csv(filename="stats.csv"):
    stats = {}
    order = []
    display_map = {}

    with open(filename, encoding='utf-8') as f:
        reader = csv.DictReader(f)
        for row in reader:
            name = row["名前"]
            form = row["形態"]
            key = (name, form)
            AS = float(row["AS"])
            BS = float(row["BS"])
            CS = float(row["CS"])
            stats[key] = (AS, BS, CS)
            order.append(key)

            if "表示" in row and row["表示"]:
                related = row["表示"].split(",")
                parsed_related = []
                for entry in related:
                    entry = entry.strip()
                    if "(" in entry and ")" in entry:
                        n = entry[:entry.index("(")]
                        f = entry[entry.index("(")+1:entry.index(")")].strip()
                    else:
                        n, f = entry.strip(), ""
                    parsed_related.append((n, f))
                display_map[key] = parsed_related

    order.sort(key=lambda x: x[0])
    return stats, order, display_map

def load_cpm_from_csv(filename="cpm.csv"):
    cpm_list = []
    cpm_to_level = {}
    level_to_cpm = {}

    with open(filename, encoding='utf-8') as f:
        reader = csv.DictReader(f)
        for row in reader:
            level = float(row["Level"])
            cpm = float(row["CPM"])
            cpm_list.append(cpm)
            cpm_to_level[round(cpm, 10)] = level
            level_to_cpm[level] = cpm

    return cpm_list, cpm_to_level, level_to_cpm

# -----------------------------
# 🧠 SCP/CP 計算関数
# -----------------------------

# SCP計算（キャッシュ非使用）
def calculate_cp(AS, BS, CS, AK, BK, CK, CPM):
    return math.floor((AS + AK) * math.sqrt(BS + BK) * math.sqrt(CS + CK) * (CPM ** 2) / 10)

def calculate_scp(AS, BS, CS, AK, BK, CK, CPM):
    hp_term = math.floor((CS + CK) * CPM)
    val = ((((AS + AK) * CPM) * ((BS + BK) * CPM) * hp_term) ** (2 / 3)) / 10
    return val

# -----------------------------
# 📊 SCP ランキング取得（タイ順位付き）（並列化あり、安全版）
# -----------------------------

import numpy as np

def get_scp_rankings_numpy(AS, BS, CS, cpm_list, cp_limit):
    iv_range = np.arange(16)
    AK, BK, CK = np.meshgrid(iv_range, iv_range, iv_range, indexing='ij')
    AK = AK.flatten()
    BK = BK.flatten()
    CK = CK.flatten()
    n = len(AK)

    AS_arr = np.full(n, AS)
    BS_arr = np.full(n, BS)
    CS_arr = np.full(n, CS)

    best_scp = np.zeros(n)
    best_cp = np.zeros(n)
    best_cpm = np.zeros(n)

    for CPM in cpm_list:
        atk = (AS_arr + AK) * CPM
        dfs = (BS_arr + BK) * CPM
        hp_term = np.floor((CS_arr + CK) * CPM)
        scp = (((atk * dfs * hp_term) ** (2/3)) / 10)

        cp = np.floor((AS_arr + AK) * np.sqrt(BS_arr + BK) * np.sqrt(CS_arr + CK) * (CPM ** 2) / 10)

        # CP 制限フィルタ
        valid = True if cp_limit is None else cp <= cp_limit

        update = (scp > best_scp) & valid
        best_scp = np.where(update, scp, best_scp)
        best_cp = np.where(update, cp, best_cp)
        best_cpm = np.where(update, CPM, best_cpm)

    result = []
    for i in range(n):
        if best_scp[i] > 0:
            result.append({
                "AK": int(AK[i]), "BK": int(BK[i]), "CK": int(CK[i]),
                "SCP": float(best_scp[i]),
                "CP": int(best_cp[i]),
                "CPM": float(best_cpm[i])
            })

    # タイ順位処理
    result.sort(key=lambda x: x["SCP"], reverse=True)
    current_rank = 1
    last_scp = None
    for i, r in enumerate(result):
        if last_scp is None or r["SCP"] != last_scp:
            current_rank = i + 1
        r["rank"] = current_rank
        last_scp = r["SCP"]

    return result

def get_competitive_rank(rankings, AK, BK, CK):
    for r in rankings:
        if r["AK"] == AK and r["BK"] == BK and r["CK"] == CK:
            return r["rank"]
    return None

# -----------------------------
# 📈 UI 入力箇所
# -----------------------------

def display_result(stats_dict, cpm_list, cpm_to_level, level_to_cpm, order, display_map):
    names = []
    seen = set()
    for name, form in order:
        if name not in seen:
            names.append(name)
            seen.add(name)

    name_dropdown = widgets.Combobox(
        placeholder='ポケモン名を入力',
        options=names,
        description='名前:',
        ensure_option=False,  # 入力した名前が候補になくてもOK
        layout=widgets.Layout(width='90%', margin='1px')
    )
    form_dropdown = widgets.Dropdown(
        description='形態:',
        layout=widgets.Layout(width='90%', margin='1px')
    )
    level_limit_dropdown = widgets.Dropdown(
        options=[("レベル50まで", 50), ("レベル51まで", 51), ("レベル55まで", 55)],
        description='レベル上限:',
        layout=widgets.Layout(width='90%', margin='1px')
    )
    range_input = widgets.BoundedIntText(
        value=100,
        min=1, max=4096,
        step=1,
        description="順位範囲:",
        layout=widgets.Layout(width='90%', margin='1px')
    )


    mega_checkbox = widgets.Checkbox(
        value=True,
        description="メガ進化を除外",
        layout=widgets.Layout(margin='1px 0 0 0')
    )
    super_league_cb = widgets.Checkbox(value=True, description="スーパーリーグを表示")
    hyper_league_cb = widgets.Checkbox(value=True, description="ハイパーリーグを表示")
    little_league_cb = widgets.Checkbox(value=True, description="リトルカップを表示")
    open_league_cb = widgets.Checkbox(value=False, description="制限なしを表示")



    line1 = widgets.HBox([name_dropdown])
    line2 = widgets.HBox(
        [form_dropdown],
        layout=widgets.Layout(align_items='center', justify_content='flex-start')
    )


    # 詳細設定画面
    detail_box = widgets.VBox([level_limit_dropdown, range_input, mega_checkbox ,super_league_cb, hyper_league_cb, little_league_cb, open_league_cb],
                             layout=widgets.Layout(margin='10px 0'))
    toggle_detail_btn = widgets.ToggleButton(
        value=False,
        description='詳細設定を表示',
        icon='eye',  # アイコンはお好みで
        layout=widgets.Layout(width='100%', margin='6px 0 0 0')
    )
    detail_box.layout.display = 'none'
    def toggle_detail_visibility(change):
        if change['new']:
            detail_box.layout.display = 'block'
            toggle_detail_btn.description = "詳細設定を隠す"
            toggle_detail_btn.icon = 'eye-slash'
        else:
            detail_box.layout.display = 'none'
            toggle_detail_btn.description = "詳細設定を表示"
            toggle_detail_btn.icon = 'eye'
    toggle_detail_btn.observe(toggle_detail_visibility, names='value')



    # 個体値ボタン
    class IVSelector:
        def __init__(self, label):
            self.label = widgets.HTML(f"<b>{label}</b>")
            self.buttons = []
            self.selected_index = 0
            self.space = widgets.Box(layout=widgets.Layout(width='10px'))
            self.container_children = []
            # ボタン作成（flexで幅均等割り振り）
            for i in range(16):
                btn = widgets.Button(
                    description=str(i),
                    layout=widgets.Layout(flex='1 1 auto', height='28px', min_width='5px', padding='0px', margin='0'),
                    button_style='',  # 色は後で変更
                    style={'font_size': '2px'}
                )
                btn.on_click(self.on_click)
                self.buttons.append(btn)
            # 子要素に空白スペースを挟む
            for i in range(16):
                self.container_children.append(self.buttons[i])  # self.buttons を使う
                if i in [0, 5, 10]:  # ここで空白を入れる
                    self.container_children.append(self.space)
            self.box = widgets.HBox(self.container_children, layout=widgets.Layout(width='100%', align_items='center'))
            self.widget = widgets.VBox([self.label, self.box])
            self.update_button_colors()  # 初期色反映
        def on_click(self, b):
            self.selected_index = int(b.description)
            self.update_button_colors()
        def update_button_colors(self):
            for i, btn in enumerate(self.buttons):
                if i < self.selected_index:
                    # 選択より左は薄い青
                    btn.style.button_color = '#ffd3a8'  # 好きな色に
                    btn.layout.border = '1px solid #7aaeff'
                    btn.layout.color = '#000000'
                elif i == self.selected_index:
                    # 選択中は濃い青
                    btn.style.button_color = '#ff7f7f'
                    btn.layout.border = '2px solid #1e3a8a'
                    btn.layout.color = '#000000'
                else:
                    # 右はデフォルト白
                    btn.style.button_color = '#999999'
                    btn.layout.border = '1px solid #ccc'
                    btn.layout.color = '#000000'
        def get_value(self):
            return self.selected_index
    ak_selector = IVSelector("攻撃個体値")
    bk_selector = IVSelector("防御個体値")
    ck_selector = IVSelector("HP個体値")
    iv_box = widgets.VBox([
        ak_selector.widget,
        bk_selector.widget,
        ck_selector.widget
    ])
    iv_box.layout = widgets.Layout(width='100%')




    # 検索ボタンのスタイル調整
    button = widgets.Button(
    description="Search",
    button_style='success',
    layout=widgets.Layout(width='100%', height='40px')
    )
    button.style.button_color = '#000000'  # ボタンの背景色
    button.style.font_weight = 'bold'
    button.style.font_size = '12px'
    button.style.text_color = 'white'
    button.layout.margin = '24px 0px 24px 0px'  # ボタン余白
    output = widgets.Output()


    def update_forms(change):
        selected_name = change["new"]
        if not selected_name:
            return  # 入力が空の場合は何もしない
        forms = [form for (n, form) in order if n == selected_name]
        if mega_checkbox.value:
            forms = [f for f in forms if not ("メガ" in f)]
        form_dropdown.options = forms
        if forms:
            form_dropdown.value = forms[0]

    name_dropdown.observe(update_forms, names='value')
    mega_checkbox.observe(lambda change: update_forms({"new": name_dropdown.value}), names='value')
    update_forms({"new": name_dropdown.value})


# -----------------------------
# 📈 UI 表示処理（HTMLカード形式）
# -----------------------------


    def on_click(b):
        name = name_dropdown.value
        form = form_dropdown.value
        AK, BK, CK = ak_selector.get_value(), bk_selector.get_value(), ck_selector.get_value()
        level_limit = level_limit_dropdown.value
        limit_rank = range_input.value

        displayed_htmls = {}

        # ✅ チェックされているリーグのみを処理対象に
        selected_leagues = []
        if super_league_cb.value:
            selected_leagues.append((1500, "スーパーリーグ"))
        if hyper_league_cb.value:
            selected_leagues.append((2500, "ハイパーリーグ"))
        if little_league_cb.value:
            selected_leagues.append((500, "リトルカップ"))
        if open_league_cb.value:
            selected_leagues.append((None, "制限なし"))

        # ⚠️ チェックが全て外れていたら警告を出す
        if not selected_leagues:
            with output:
                clear_output(wait=True)
                display(HTML("<p style='color:red;'>⚠️ 表示するリーグが選択されていません。</p>"))
            return
        with output:
            clear_output(wait=True)
            if (name, form) not in stats_dict:
                display(HTML("<p style='color:red;'>⚠️ 種族値データが見つかりません</p>"))
                return

            related_keys = display_map.get((name, form), [])
            all_keys = list(reversed(related_keys))
            final_keys_ordered = [k for k in all_keys if k in stats_dict and (not mega_checkbox.value or "メガ" not in k[1])]
            allowed_cpms = [cpm for cpm in cpm_list if cpm_to_level.get(round(cpm, 10), 0) <= level_limit]

            league_colors = {
                "スーパーリーグ": ("#b22222", "#00008b"),
                "ハイパーリーグ": ("#f0e68c", "#444444"),
                "リトルカップ": ("#da81b2", "#fdede4"),
                "制限なし": ("#c71585", "#8b008b")
            }

            for key in final_keys_ordered:
                name_, form_ = key
                AS, BS, CS = stats_dict[key]

                for cp_limit, label in selected_leagues:
                    rankings = get_scp_rankings_numpy(AS, BS, CS, allowed_cpms, cp_limit)
                    outer_color, inner_color = league_colors.get(label, ("#999999", "#cccccc"))
                    target_rank = get_competitive_rank(rankings, AK, BK, CK)

                    if target_rank is not None:
                        result = next(r for r in rankings if r['AK'] == AK and r['BK'] == BK and r['CK'] == CK)
                        scp = result['SCP']
                        cp = result['CP']
                        level = cpm_to_level.get(round(result['CPM'], 10), '?')
                        total = len(rankings)
                        scp_max = rankings[0]['SCP']
                        scp_min = rankings[-1]['SCP']
                        top = rankings[0]
                        sub_range = rankings[:limit_rank]
                        min_vals = {k: min(r[k] for r in sub_range) for k in ['AK', 'BK', 'CK']}
                        max_vals = {k: max(r[k] for r in sub_range) for k in ['AK', 'BK', 'CK']}
                        # 1位個体値情報を取得
                        top_ak, top_bk, top_ck = top['AK'], top['BK'], top['CK']
                        if isinstance(level, (int, float)) and level >= 50:
                            if top_ak == 15 and top_bk == 15 and top_ck == 15:
                                bg_color = "#aaaaaa"  # ALL15かつLv50超 → 背景色
                            else:
                                bg_color = "#bfbfbf"  # Lv50超 → 背景色
                        else:
                            if target_rank == 1:
                                bg_color = "#ffffcc"  # 1位:
                            elif target_rank <= 10:
                                bg_color = "#ffffe0"  # 2〜10位:
                            elif target_rank <= 50:
                                bg_color = "#ffffef"  # 11〜50位:
                            else:
                                bg_color = "#f9f9f9"  # 51位以降:
                        # Lv表示を黄色にするかどうか
                        level_color = "#FFFF00" if isinstance(level, (int, float)) and level >= 50 else "#666"
                        # 1位個体値を黄色にするかどうか（ALL15かつLv50以上）
                        iv_color = "#FFFF00" if (isinstance(level, (int, float)) and level >= 50 and top['AK'] == 15 and top['BK'] == 15 and top['CK'] == 15) else "#444"


                        card_html = f"""
                        <div style='position: relative; width: 100%; max-width: 300px; box-sizing: border-box; background: {bg_color}; border-radius: 12px; padding: 12px; margin: 8px; border: 1px solid #ccc; box-shadow: 2px 2px 6px rgba(0,0,0,0.1); line-height: 1.6;'>

                          <!-- 🔺 右上にリーグ色の二重三角形 -->
                          <div style='
                            position: absolute;
                            top: 0;
                            right: 0;
                            width: 0;
                            height: 0;
                            border-top: 40px solid {outer_color};
                            border-left: 40px solid transparent;
                            border-radius: 0 12px 0 0;
                            z-index: 10;
                          '></div>
                          <div style='
                            position: absolute;
                            top: 0;
                            right: 0;
                            width: 0;
                            height: 0;
                            border-top: 35px solid {inner_color};
                            border-left: 35px solid transparent;
                            border-radius: 0 10px 0 0;
                            z-index: 11;
                          '></div>
                        """

                        card_html += (
                            f"<div style='font-weight: bold; color: #444;'>"
                            f"<span style='font-size: 1.2em; display: inline-block; vertical-align: bottom;'>🏆 {label}順位:</span> "
                            f"<span style='font-size: 1.2em; color: #aa3333; display: inline-block; vertical-align: bottom;'> {target_rank}</span>"
                            f"</div>"
                        )

                        # SCP上位％を使ったゲージ表示（順位のすぐ下）
                        percent = target_rank / total
                        percent_display = round(percent * 100, 2)

                        # 色分けルールに基づいて色決定
                        if target_rank <= 100:
                            bar_color = "#00ff00"  # 100位
                        elif percent <= 0.5:
                            bar_color = "#aadd88"  # 5割
                        elif percent <= 0.8:
                            bar_color = "#d0af4c"  # 8割
                        else:
                            bar_color = "#cd5c5c"  # （下位）

                        # ゲージの塗り幅（100%:空 / 0%:満タン）
                        filled_width = int((1.0 - percent) * 100)

                        card_html += f"""
                        <div style="margin: 4px 0 8px 0;">
                          <div style="width: 100%; height: 10px; background: #eee; border-radius: 5px; overflow: hidden; box-shadow: inset 0 1px 2px rgba(0,0,0,0.1);">
                            <div style="width: {filled_width}%; height: 100%; background: {bar_color};"></div>
                          </div>
                        </div>
                        """

                        card_html += f"<div style='font-size: 0.50em; color: #666;'>上位 {round(target_rank / total * 100, 2)}% ／ SCP:{math.floor(scp)} ／ CP:{cp} ／ <span style='color:{level_color}'>Lv:{level}</span></div>"
                        card_html += f"<div style='color: #aa3333;'>最大との差: {math.floor(scp_max) - math.floor(scp):+} ／ 最小との差: {math.floor(scp_min) - math.floor(scp):+}</div>"


                        try:
                            idx = final_keys_ordered.index((name_, form_))# まず今のポケモンのindexをfinal_keys_orderedで取得
                        except ValueError:
                            idx = None
                        prev_keys = final_keys_ordered[idx+1:] if idx is not None else []# 名前より順番が前のポケモンリスト（逆順リストなので idxより後ろ）
                        prev_cp_html = "<div style='font-size:0.50em; color:#555;'>"
                        for pk in prev_keys:
                            if pk not in stats_dict:
                                continue
                            AS_p, BS_p, CS_p = stats_dict[pk]
                            cp_val = calculate_cp(AS_p, BS_p, CS_p, AK, BK, CK, result['CPM'])
                            prev_cp_html += f"{pk[0]} {pk[1]}: 上限CP{cp_val}（Lv:{level}）<br>"# 同じ条件（個体値AK,BK,CK、CPM）でCP計算して表示
                        prev_cp_html += "</div>"
                        card_html += prev_cp_html



                        card_html += "<hr style='margin: 6px 0; border: none; border-top: 1px solid #ccc;'>"
                        card_html += f"<div style='font-size: 0.70em; color: {iv_color};'>1位個体値 → 攻:{top['AK']} 防:{top['BK']} HP:{top['CK']}</div>"
                        card_html += f"<div style='font-size: 0.70em; color: #444;'>個体値範囲（{limit_rank}位内）→ 攻:{min_vals['AK']}〜{max_vals['AK']} 防:{min_vals['BK']}〜{max_vals['BK']} HP:{min_vals['CK']}〜{max_vals['CK']}</div>"
                    else:
                        card_html += f"<div>🏆 {label}<br>対象個体値はCP超過で無効</div>"

                    card_html += "</div>"

                    key_for_group = (name_, form_)
                    if key_for_group not in displayed_htmls:
                        displayed_htmls[key_for_group] = []
                    displayed_htmls[key_for_group].append(card_html)

                with output:
                    clear_output(wait=True)
                    for (n, f), cards in displayed_htmls.items():
                        title = f"{n}/{f}" if f else n
                        combined_html = f"<div style='margin-bottom: 24px;'>"
                        combined_html += f"<h2 style='text-align: center; margin: 8px 0 4px 0;'>{title}</h2>"
                        combined_html += f"<div style='display: flex; flex-wrap: wrap; justify-content: center;'>"
                        combined_html += "".join(cards)
                        combined_html += "</div></div>"
                        display(HTML(combined_html))

    button.on_click(on_click)

    display(line1)
    display(line2)
    display(toggle_detail_btn)  # ← ここでボタン表示
    display(detail_box)         # ← 非表示スタート
    display(iv_box)
    display(button)
    display(output)

# -----------------------------
# ▶ 実行開始
# -----------------------------

stats_dict, order, display_map = load_stats_from_csv("stats.csv")
cpm_list, cpm_to_level, level_to_cpm = load_cpm_from_csv("cpm.csv")

display_result(stats_dict, cpm_list, cpm_to_level, level_to_cpm, order, display_map)
```
